```python
import numpy as np
from ase.build import bulk
from ase.optimize.bfgs import BFGS
from ase.constraints import UnitCellFilter
from gpaw import GPAW, FermiDirac
from gpaw import PW
from gpaw.xc.libvdwxc import vdw_df
import ase.io.cif
from ase.io import read
from ase.visualize import view
from ase.visualize.plot import plot_atoms

twolayer = read('2_MoS2.cif')
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(1,1,1)
plot_atoms(twolayer,radii=0.2, scale=3, rotation=('100x,0y,0z'))
ax.set_axis_off()
plt.savefig('structure.png')
```

```python
calc = GPAW(xc=vdw_df(),
            mode=PW(600),
            occupations=FermiDirac(5e-3),
            kpts=(4, 4, 1),
            )

twolayer.set_calculator(calc)

#twolayer.get_potential_energy()

relax = BFGS(twolayer)
relax.run(fmax=0.01)
```
